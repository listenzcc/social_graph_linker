```python
import random
import pandas as pd
```

```python
departments = ['Organization', 'Police', 'Propagation', 'Military', 'Law', 'Industry']

def random_major(majors=departments):
    return random.choice(majors)

def random_name(region=None, k=None):
    if region is None:
        region = [chr(e) for e in range(ord('a'), ord('z')+1)]
    if k is None:
        k = [random.randint(3, 5), random.randint(5, 10)]
    xing = ''.join(random.choices(region, k=k[0])).title()
    ming = ''.join(random.choices(region, k=k[1])).title()
    return ' '.join([xing, ming])

def random_age(low=40, high=75):
    return random.randint(low, high)

def level3_uid(idx):
    return 'L3-{idx:04d}'.format(idx=idx)
```

```python
num_level3 = 100
```

```python
level3_df = pd.DataFrame(columns=['name', 'major', 'age', 'uid'])
level3_df
```

```python
for _ in range(num_level3):
    uid = level3_uid(_)
    info = dict(
        major = random_major(),
        name = random_name(),
        age = random_age(),
        uid = uid
    )
    level3_df = level3_df.append(pd.Series(info, name=uid))
level3_df
```
